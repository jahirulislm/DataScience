# Data Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# data load
df = sns.load_dataset('tips')
df.head()
```

```python
# Draw a histplot kde on it
plt.figure(figsize=(8,5))
sns.histplot(data= df, x = 'total_bill', kde=True)
```

```python
# Draw a histplot kde on it
plt.figure(figsize=(8,5))
sns.histplot(data= df, x = 'tip', kde=True)
```

```python
# Draw a histplot kde on it
plt.figure(figsize=(8,5))
sns.histplot(data= df, x = 'day', kde=True)
```

```python
# Draw a histplot kde on it
plt.figure(figsize=(8,5))
sns.histplot(data= df, x = 'size', kde=True)
```

```python
# titanic
df_1 = sns.load_dataset('titanic')
df_1.head()
```

```python
sns.histplot(data= df_1, x='fare', kde=True, bins=50, hue='class')
```

```python
sns.boxplot(data =df_1, x='age', hue='who')
```
