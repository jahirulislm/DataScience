```python
df = sns.load_dataset('titanic')
df.info()
```

```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(12,6)) 
# sns.scatterplot(data=df, x= 'age', y='fare', hue='sex')

# style a plot
sns.set_theme(style='darkgrid')

# change the scatter points color
palette = {'male': 'yellow', 'female':'red'}
sns.scatterplot(data=df, x= 'age', y='fare', hue='sex',size='age', palette=palette)

# saving a plot, keep the dpi under 300dpi, bit heavy 600dpi, normally 150dpi, format- png
# plt.savefig('./output/scatter_plot.png', dpi=300)
plt.savefig('./output/scatter_plot.svg', dpi=300)
```

```python
df['class'].unique()
```

```python
# Box plot
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(12,6)) 
# sns.scatterplot(data=df, x= 'age', y='fare', hue='sex')

# style a plot
sns.set_theme(style='darkgrid')

# change the scatter points color
# palette = {'male': 'yellow', 'female':'red'}

#based on class
palette = {'First': 'green', 'Second':'red','Third':'blue'}
sns.boxplot(data=df, x= 'sex', y='age', hue='class', palette=palette)

# saving a plot, keep the dpi under 300dpi, bit heavy 600dpi, normally 150dpi, format- png
# plt.savefig('./output/scatter_plot.png', dpi=300)
plt.savefig('./output/scatter_plot.svg', dpi=300)

# change legend position and title
plt.legend(loc='upper left', title='Class Name')
```

Interpretation:
    1. we have age of passengers at x-axis and fare at y-axis
    2. The scatter plot shows the relationship between age and fare of titanic passengers.
    3. We further grouped the passenger by their genders such as male(green point) and female( red point)
    4. Two male passengers gave the highest fare in titanic, and they might be travelling first class.
    5. The oldest person at titanic is also a male of 80 years old (as shown in the scatter plot).


```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(12,6)) 
# sns.scatterplot(data=df, x= 'age', y='fare', hue='sex')

# style a plot
sns.set_theme(style='darkgrid')

# change the scatter points color
palette = {'First': 'green', 'Second':'red','Third':'blue'}
sns.boxenplot(data=df, x= 'sex', y='age', hue='class', palette=palette)

# saving a plot, keep the dpi under 300dpi, bit heavy 600dpi, normally 150dpi, format- png
# plt.savefig('./output/scatter_plot.png', dpi=300)
# plt.savefig('./output/scatter_plot.svg', dpi=300)

# change the x-tick label names
plt.xticks(ticks=[0,1],labels=['Male','Female'])
```

```python
# make scatter plot using plotly
import plotly.express as px

fig= px.scatter(df, x='age',y='fare', color='sex')
fig.show()
```
